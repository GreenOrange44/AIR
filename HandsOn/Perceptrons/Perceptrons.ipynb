```python
#CREATING A PERCEPTRON WORKING ON MULITPLE WEIGHTED INPUTS AND ONE BIAS

import numpy as np


# Define the activation function (e.g., sigmoid function)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


class Neuron:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias


    # Function for calculating the output of the neuron
    def feedforward(self, inputs):
        # Weighted sum of inputs / weights + bias
        total = np.dot(self.weights, inputs) + self.bias
        return sigmoid(total)


# Define weights, bias and inputs
weights = np.array([0.3, 0.1, 0.1])  # considering 3 input
bias = 1
inputs = np.array([1, -1, 1])


# Initialize a neuron
neuron = Neuron(weights, bias)


# Get the output of the neuron
output = neuron.feedforward(inputs)
print(output)



```

```python
#Making a graph of the output returned by sigmoid activated neuron over a range of input

import numpy as np
import matplotlib.pyplot as plt


def sigmoid(x):
    return 1 / (1 + np.exp(-x))


class Neuron:
    def __init__(self, weight, bias):
        self.weight = weight
        self.bias = bias


    def feedforward(self, input):
        total = self.weight * input + self.bias
        return sigmoid(total)


# Define weight, bias
weights = np.array([0.3])  # considering 1 input
bias = 1


# Initialize a neuron
neuron = Neuron(weights, bias)


# Define a range of inputs
inputs = np.linspace(-100, 100, 1000)


# Compute the output for each input
outputs = [neuron.feedforward(i) for i in inputs]


# Plot the output
plt.figure(figsize=(8,6))
plt.plot(inputs, outputs)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Output of the neuron for various inputs")
plt.grid(True)
plt.show()





```

# Perceptrons that learn itself

```python
import numpy as np


# Define the Perceptron Class
class Perceptron(object):
    def __init__(self, no_of_inputs, epochs=100, learning_rate=0.01):
        self.epochs = epochs
        self.learning_rate = learning_rate
        self.weights = np.zeros(no_of_inputs + 1) # Add 1 for bias weight


    def predict(self, inputs):


        summation = np.dot(inputs, self.weights[1:]) + self.weights[0]
        if summation > 0:
          activation = 1
        else:
          activation = 0
        return activation


    def train(self, training_inputs, labels):
        for _ in range(self.epochs):
            print(f"Iteration {_}: Bias and weight: {self.weights}")
            for inputs, label in zip(training_inputs, labels):
                prediction = self.predict(inputs)
                self.weights[1:] += self.learning_rate * (label - prediction) * inputs #weights
                self.weights[0] += self.learning_rate * (label - prediction) #bias




# Initialize the single-input Perceptron
perceptron = Perceptron(1)
X = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = np.array([0, 0, 1, 1, 0, 0, 1, 1])


# Reshape X to match the Perceptron input shape
X = np.reshape(X, (X.shape[0], -1))
print("hi", X)

# Train the Perceptron
perceptron.train(X, y)




def test_perceptron():
    user_input = input("Please enter an input: ")
    input_value = np.array([float(user_input)])
    prediction = perceptron.predict(input_value)
    print(f"The predicted output is: {prediction}")


test_perceptron()







```
