```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('palm.png')

h, w = img.shape[:2]

start_row, start_colm = int(h*0.45), int(w*0.35)
end_row, end_colm = int(h*0.76), int(w*0.84)


```

```python
cropped_img = img[start_row:end_row, start_colm:end_colm]
#cv2.imshow('ROI', cropped_img)
plt.imshow(cropped_img)
plt.show()
```

```python
cv2.waitKey()
cv2.destroyAllWindows()

cv2.imwrite("roi.jpg", cropped_img)
```

# **Bilateral Filter**

```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread("palm.png")
dimpixel=7
sigmaColor=100
sigmaSpace=100
img_bil=cv2.bilateralFilter(img, dimpixel, sigmaColor, sigmaSpace)
#cv2.imshow('original', img)
plt.imshow(img_bil)
plt.show()

#cv2.imshow("filter", img_bil)
plt.imshow(img_bil)
plt.show()
cv2.waitKey()
cv2.destroyAllWindows()
```

# **Sobel Filter**

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage.filters import sobel

img=cv2.imread("palm.png")

img_sobel=sobel(img)
#cv2.imshow('original', img)
plt.imshow(img_sobel)
plt.show()

#cv2.imshow("Sobel Image", img_sobel)
plt.imshow(img_sobel)
plt.show()

cv2.waitKey()
cv2.destroyAllWindows()
```

# **Otsu Threshold**

```python


import cv2
import numpy as np
from matplotlib import pyplot as plt

img=cv2.imread("palm.png", 0)

ret, th=cv2.threshold(img, 0, 255, cv2.THRESH_OTSU)

ret, bn = cv2.threshold(img, 127,255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU) # to invert the OTSU

#cv2.imshow('original', img)
plt.imshow(img)
plt.show()

#cv2.imshow("threshold", th)
plt.imshow(th)
plt.show()


#cv2.imshow("Invert binary", bn)
plt.imshow(bn)
plt.show()

cv2.waitKey()
cv2.destroyAllWindows()
```

# **Gausian Blur**

```python

import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread("palm.png")

blur_img=cv2.GaussianBlur(img, (15,15), sigmaX=34, sigmaY=36)
#cv2.imshow('original', img)
plt.imshow(img)
plt.show()

#cv2.imshow("Gaussian Blur", blur_img)
plt.imshow(blur_img)
plt.show()

cv2.waitKey()
cv2.destroyAllWindows()
```

# **Laplacian Fitler**

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread("palm.png")

lap_img=cv2.Laplacian(img, cv2.CV_64F, ksize=3)
#cv2.imshow('original', img)
plt.imshow(img)
plt.show()

#cv2.imshow("Laplacian", lap_img)
plt.imshow(lap_img)
plt.show()

cv2.waitKey()
cv2.destroyAllWindows()
```

# **Laplacian on top of Gausian Blur**

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread("palm.png")

blur_img=cv2.GaussianBlur(img, (3,3), sigmaX=34, sigmaY=36)


cv2.waitKey()
cv2.destroyAllWindows()


lap_img=cv2.Laplacian(blur_img, cv2.CV_64F, ksize=3)
#cv2.imshow('original', img)
plt.imshow(img)
plt.show()

#cv2.imshow("Laplacian", lap_img)
plt.imshow(lap_img)
plt.show()

cv2.waitKey()
cv2.destroyAllWindows()

```

# **Canny Edge**

```python

import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread("palm.png")

can_img=cv2.Canny(img, 30, 30)
#cv2.imshow('original', img)
plt.imshow(img)
plt.show()

#cv2.imshow("Canny Image", can_img)
plt.imshow(can_img)
plt.show()

cv2.waitKey()
cv2.destroyAllWindows()
```

# **Low Pass Filter**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread("palm.png")

kernel=np.ones((5,5), np.float32)/80
lpf_img=cv2.filter2D(img, -1, kernel) # -1 is the depth
#cv2.imshow("original", img)
plt.imshow(img)
plt.show()

#cv2.imshow("LPF img", lpf_img)
plt.imshow(lpf_img)
plt.show()

cv2.waitKey()
cv2.destroyAllWindows()
```
