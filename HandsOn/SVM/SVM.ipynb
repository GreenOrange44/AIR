```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd


# Step 1: Load the dataset
data = pd.read_csv('SVM_d1.csv')
X = data[['X1', 'X2']].values  # We will use the 'X1' and 'X2' features
y = data['y'].values
```

```python
# Step 2: Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


```

```python
# Step 3: Train the SVM model
# We'll use a linear kernel for simplicity
svm = SVC(kernel='linear', C=1.0)
svm.fit(X_train, y_train)


```

```python
# Step 4: Make predictions
y_pred = svm.predict(X_test)


```

```python
# Step 5: Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
print('Classification Report:')
print(classification_report(y_test, y_pred))


```

```python
# Identify support vectors
support_vectors = svm.support_vectors_
print(f'Support Vectors:\n{support_vectors}')


```

```python
# Step 6: Visualize the decision boundaries and support vectors
# Create a mesh to plot the decision boundaries
h = .02  # step size in the mesh
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))


# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])


# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.coolwarm)


# Plot also the training points and support vectors
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.coolwarm, label='Data Points')
plt.scatter(support_vectors[:, 0], support_vectors[:, 1], edgecolors='k', facecolors='none', s=100, linewidths=1.5, label='Support Vectors')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('SVM Decision Surface with Support Vectors')
plt.legend()


# Plot the margin lines
w = svm.coef_[0]
a = -w[0] / w[1]
xx_margin = np.linspace(x_min, x_max)
yy_margin = a * xx_margin - (svm.intercept_[0]) / w[1]
yy_margin_down = yy_margin - 1 / w[1]
yy_margin_up = yy_margin + 1 / w[1]


plt.plot(xx_margin, yy_margin, 'k--', label='Decision Boundary')
plt.plot(xx_margin, yy_margin_down, 'k--', label='Margin')
plt.plot(xx_margin, yy_margin_up, 'k--', label='Margin')
plt.legend()
plt.show()


```
